```python
import os
import glob
import shutil

# root_path->
root_path = 'SeaDronesSee/SeaDronesSee'
# newpath->
newpath = 'SeaDronesSee/SeaDronesSee_new'
# 
ortrimg = root_path + '/images/train'
orvalimg = root_path + '/images/val'
ortrlabel = root_path + '/labels/train'
orvallabel = root_path + '/labels/val'

# 
os.mkdir(newpath)
os.mkdir(newpath + '/labels')
os.mkdir(newpath + '/images')
os.mkdir(newpath + '/images/train')
os.mkdir(newpath + '/labels/train')
os.mkdir(newpath + '/images/val')
os.mkdir(newpath + '/labels/val')

# 
x = 1


def getimgpath(path):
    """

    :param path: 
    :return:  
    :notice:  count % 10 == x  x 
    """

    # file_list = glob.glob(ortrimg + '/*.jpg')
    file_list = glob.glob(path + '/*.jpg')
    nlist = []
    count = 0
    for i in file_list:
        # print(i)
        # 9
        if count % 10 == x:
            nlist.append(i)
        count += 1
    # print(file_list, '\n', nlist)
    # nlist 
    return file_list, nlist


def getlabelpath(path):
    """
     getimgpath(path)
    :param path:
    :return:
    """
    file_list = glob.glob(path + '/*.txt')
    nlist = []
    count = 0
    for i in file_list:
        # print(i)
        if count % 10 == x:
            nlist.append(i)
        count += 1
    # print(file_list, '\n', nlist)
    return file_list, nlist
    return


def getTnV():
    """
    
    :return:
    """

    train, train_new = getimgpath('SeaDronesSee/SeaDronesSee/images/train')
    val, val_new = getimgpath('SeaDronesSee/SeaDronesSee/images/val')
    print(train_new, '\n', val_new)
    tra_label, tralabel_new = getlabelpath('SeaDronesSee/SeaDronesSee/labels/train')
    val_label, vallabel_new = getlabelpath('SeaDronesSee/SeaDronesSee/labels/val')
    print(tralabel_new, '\n', vallabel_new)

    return train_new, tralabel_new, val_new, vallabel_new


def makenew():
    """
    
    :return:
    """

    oldfile = getTnV()
    ptrain = newpath + '/images/train/'
    pval = newpath + '/images/val/'
    lptrain = newpath + '/labels/train/'
    lpval = newpath + '/labels/val/'
    newfl = [ptrain, lptrain, pval, lpval]
    for i in oldfile:
        print(i)

    for i in range(len(oldfile)):
        for j in oldfile[i]:
            print(j, ' 2 ', newfl[i])
            shutil.copy(j, newfl[i])


if __name__ == "__main__":
    makenew()
```
