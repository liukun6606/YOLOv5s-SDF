```python
import os
import glob
import shutil

# root_path->旧的数据集目录
root_path = '/SeaDronesSee'
# newpath->新的数据集目录
newpath = '/SeaDronesSee_new'
# 数据集旧目录对应文件目录
ortrimg = root_path + '/images/train'
orvalimg = root_path + '/images/val'
ortrlabel = root_path + '/labels/train'
orvallabel = root_path + '/labels/val'

# 创建新数据集文件目录树
os.mkdir(newpath)
os.mkdir(newpath + '/labels')
os.mkdir(newpath + '/images')
os.mkdir(newpath + '/images/train')
os.mkdir(newpath + '/labels/train')
os.mkdir(newpath + '/images/val')
os.mkdir(newpath + '/labels/val')

# 数据集文件选取控制变量
x = 9


def getimgpath(path):
    """

    :param path: 图片文件目录
    :return: 旧文件目录列表和 筛选后新列表
    :notice: 文件选取控制由 count % 10 == x 来控制，保证每十个文件选取第 x 个，放入新的数据集
    """

    # file_list = glob.glob(ortrimg + '/*.jpg')
    file_list = glob.glob(path + '/*.jpg')
    nlist = []
    count = 0
    for i in file_list:
        # print(i)
        # 数据选取控制，当前选取第9个
        if count % 10 == x:
            nlist.append(i)
        count += 1
    # print(file_list, '\n', nlist)
    # nlist 为新的数据列表
    return file_list, nlist


def getlabelpath(path):
    """
    参考 getimgpath(path)
    :param path:
    :return:
    """
    file_list = glob.glob(path + '/*.txt')
    nlist = []
    count = 0
    for i in file_list:
        # print(i)
        if count % 10 == x:
            nlist.append(i)
        count += 1
    # print(file_list, '\n', nlist)
    return file_list, nlist
    return


def getTnV():
    """
    重新组织目标数据集文件目录关系
    :return:
    """

    train, train_new = getimgpath('/SeaDronesSee/images/train')
    val, val_new = getimgpath('/SeaDronesSee/images/val')
    print(train_new, '\n', val_new)
    tra_label, tralabel_new = getlabelpath('/SeaDronesSee/labels/train')
    val_label, vallabel_new = getlabelpath('/SeaDronesSee/labels/val')
    print(tralabel_new, '\n', vallabel_new)

    return train_new, tralabel_new, val_new, vallabel_new


def makenew():
    """
    将待选数据，全部复制到新的数据集目录中
    :return:
    """

    oldfile = getTnV()
    ptrain = newpath + '/images/train/'
    pval = newpath + '/images/val/'
    lptrain = newpath + '/labels/train/'
    lpval = newpath + '/labels/val/'
    newfl = [ptrain, lptrain, pval, lpval]
    for i in oldfile:
        print(i)

    for i in range(len(oldfile)):
        for j in oldfile[i]:
            print(j, ' 2 ', newfl[i])
            shutil.copy(j, newfl[i])


if __name__ == "__main__":
    makenew()
```
